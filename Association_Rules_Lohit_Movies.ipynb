# Visualising Rules paramters with diff plots

### Scatter plot

```python
fig,ax=plt.subplots(1,2, figsize=(12,8))

sns.scatterplot(ax=ax[0],x='antecedent support',y='consequent support',hue='confidence',data=rules_1)
ax[0].set_title('Confidence')

sns.scatterplot(ax=ax[1],x='antecedent support',y='consequent support',hue='lift',data=rules_1)
ax[1].set_title('Lift')
```

    -The first subplot shows relationship of antecedent and consequent support with confidence

### Heatmap

# Keeping confidence as constant and Support as variable

```python

confidence=[]
rule=[]
support=[]
frequent_data=[]
min_threshold=0.1
l=0

for i in np.arange(0.01,0.6,0.05):
    data=apriori(mv,min_support=i,use_colnames=True)
    rules=association_rules(data,metric='confidence',min_threshold=min_threshold)
    len_rules=len(rules)
    len_datasets=len(data)
    rule.append(len_rules)
    confidence.append(min_threshold)
    frequent_data.append(len_datasets)
    support.append(i)
    l=l+1
    
print('number of iteration :', l)
```

```python
final={'support': support,'frequent_dataset':frequent_data, 'rule': rule,'confidence ': confidence}
final=pd.DataFrame(final)
final.head()
```

```python
final.shape
```

```python
plt.plot('support','rule',data=final)
plt.xlabel("support")
plt.ylabel("rule")
plt.title("support vs rules")
```

# Keeping Support as constant and Confidence as variable (Metric- Confidence)

```python
support=[]
dataset=[]
rules=[]
confidence=[]
min_support=.01
l=0

for i in np.arange(.1,1,.01):
    frequent_dataSets = apriori(mv, min_support=min_support, use_colnames=True)
    rules1 = association_rules(frequent_dataSets, metric="confidence", min_threshold=i)
    len_datasets=len(frequent_dataSets)
    dataset.append(len_datasets)
    len_rules=len(rules1)
    rules.append(len_rules)
    confidence.append(i)
    support.append(min_support)
    l=l+1
    
print('number of iteration :', l)
```

```python
final={'support':support,'frequent_dataset':dataset,'rules':rules,'confidence':confidence}
df1=pd.DataFrame(final)
df1.head(10)
```

```python
df1.shape
```

```python
plt.figure(figsize=(12,8))


plt.plot('confidence','rules',data=df1)
plt.xlabel("confidence")
plt.ylabel("rules")
plt.title("confidence vs rules:Metric: Confidence")
```

# Changing max length in apriori algorithm and observing the changes in association rules for confidence metric

```python
data=[]
rules=[]
length=[]
confidence=[]

for l in range(1,10):
    frequent_data = apriori(mv, max_len=l,min_support=0.01, use_colnames=True)
    rules2 = association_rules(frequent_data, metric="confidence", min_threshold=0.2)
    len_data=len(frequent_data)
    data.append(len_data)
    len_rules=len(rules2)
    rules.append(len_rules)
    length.append(l)
    support.append(min_support)
    
    
print("Number of iterations:",l)
```

```python
final={'frequent_dataset':data,'rules':rules,'length':length}
df3=pd.DataFrame(final)
df3.head(10)
```

```python
plt.figure(figsize=(12,8))

sns.lineplot(data=df3,x="length",y="rules")
plt.title("length vs rules")
plt.show()
```

```python
data=[]
rules=[]
length=[]
confidence=[]

# setting the lenth at 7 as after 5 its a flat line 
for l in range(1,7):
    frequent_data = apriori(mv, max_len=l,min_support=0.01, use_colnames=True)
    rules2 = association_rules(frequent_data, metric="lift", min_threshold=0.2)
    len_data=len(frequent_data)
    data.append(len_data)
    len_rules=len(rules2)
    rules.append(len_rules)
    length.append(l)
    support.append(min_support)
    
    
print("Number of iterations:",l)
```

```python
final={'frequent_dataset':data,'rules':rules,'length':length}
df4=pd.DataFrame(final)
df4.head(10)
```
